```python
import requests
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
import numpy as np
```

Recently, I have taken up an interest in the game Pokemon Unite (available for both iphone and android)

This is a mini webscraping project to learn the basic of webscraping.
The goal is to extract all the pokemon names as well as the individual links

```python
response = requests.get('https://uniteapi.dev/db/pokemon')
print(response.status_code)
```

```python
page = response.text
#Identifying that we have manage to scrape successfully
print(page[:200])
```

```python
#Parsing our text 
soup = bs(page, 'html.parser')
#beautifying it to  put each html on its own line
print(soup.prettify())
```

```python
search = soup.find_all('a')
#We will first attempt to extract 1 pokemon: Gardevoir
print('Gardevoir link')
print(search[38])

#extracting the link for the pokemon Gardevoir
gardevoir = search[38]
```

```python
pokemon_links = [s.get('href') for s in search[25:]]
```

```python
all_text = soup.get_text()
all_text
```

```python
#identify how many pokemons are there in this page
#detect the position of the first pokemon vensaur
list(all_text).index('V')
all_text[675:690]
```

```python
#We will use regex to match all the pokemon names
pokemon_names =  (re.findall('(?im)\n+([a-z]*.*[a-z])',all_text[675:]))
print(pokemon_names[:20])
print('*'*30)
print(pokemon_names[-9:])
print(f'Total Number of pokemon in game: {len(pokemon_names)}')
```

```python
# I also notice the links has some duplicates
pokemon_links = ['https:/uniteapi.dev'+p for p in sorted(list(set(pokemon_links)))]
```

```python
df =  pd.DataFrame({'Pokemon':sorted(pokemon_names),'Links':sorted(pokemon_links)})
```
