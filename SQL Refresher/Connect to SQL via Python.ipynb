```python
import psycopg2 as pg2
import pandas as pd
```

# Prep to connect psycog2 to SQL

```python
#connect to postgresql
conn = pg2.connect(database = 'foodpanda',
                    user = 'postgres', #default username,
                    password = 'password')
```

```python
#Activate cursor
cur = conn.cursor()
```

## SQL queries

```python
#Movie
cur.execute('SELECT * FROM movie')
movie  = cur.fetchall()

#Rating
cur.execute('SELECT * FROM rating')
rating  = cur.fetchall()

#Reviewer
cur.execute('SELECT * FROM reviewer')
reviewer = cur.fetchall()
```

## Reading SQL Queries directly into pandas
instead of executing it and subsequently converting the list of tuples into dataframe.
We can use pandas read_sql_query function to read in directly into a dataframe

```python
#Read in SQL query directly
movie_df = pd.read_sql('SELECT * FROM movie',conn)

rating_df =  pd.read_sql('SELECT * FROM rating',conn)

reviewer_df =  pd.read_sql('SELECT * FROM reviewer',conn)
```
